right triangle

```python
n = 5 # You can change the number of rows

for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")
    print()
```

pyramid pattern

```python
n = 5 # You can change the number of rows

for i in range(1, n + 1):
    # Print leading spaces
    for j in range(n - i):
        print(" ", end="")
    # Print stars
    for k in range(2 * i - 1):
        print("*", end="")
    print()
```
